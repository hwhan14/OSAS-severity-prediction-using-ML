```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('../data/ISI,ESS,PSGdata_20220113.xlsx')
df.head()
```

# 데이터: 설문 결과 없음

```python
col_list = list(df.columns)

ISI_cols =  [s for s in col_list if 'ISI' in s]
ESS_cols =  [s for s in col_list if 'ESS' in s]
SSQ_cols =  [s for s in col_list if 'SSQ' in s]

print(ISI_cols, ESS_cols, SSQ_cols)
```

```python
for col in ISI_cols:
    print(col, '\t', len(df[df[col] == 999]))
```

```python
for col in ESS_cols:
    print(col, '\t', len(df[df[col] == 999]))
```

```python
for col in SSQ_cols:
    print(col, '\t', len(df[df[col] == 999]))
```

```python
survey_cols = ISI_cols + ESS_cols + SSQ_cols
survey_cols.remove('ISI')
survey_cols.remove('ESS')
print(survey_cols)
```

```python
no_survey = df.drop(survey_cols, axis=1)
print(no_survey.shape)
no_survey.head()
```

```python
no_survey.to_csv('../data/no_survey_data.csv', index=False)
```

# 데이터: 설문 결과 ISI

```python
ISI_df = df.drop(to_remove_cols, axis=1)
print(ISI_df.shape)
ISI_df.head()
```
