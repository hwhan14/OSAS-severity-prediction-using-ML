```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('../data/ISI,ESS,PSGdata_20220113.xlsx')
df.head()
```

```python
df.shape
```

```python
print(df.isnull().sum())
```

```python
missings = df.isnull().sum()
missings_cols = []
for col, value in missings.iteritems():
    if value != 0:
        print(col, "\t", value)
        missings_cols.append(col)
```

```python
missings_cols
```

# Data B

```python
df_b = df.drop(missings_cols, inplace=False, axis=1)
df_b.head()
```

```python
df_b.shape
```

```python
idxs_list = {}
for col in df_b.columns:
    idxs = df_b[df_b[col] == 999].index.values.astype(int)
    if len(idxs) != 0:
        idxs_list[col] = idxs
```

```python
df_b.drop(list(idxs_list.keys()), inplace=True, axis=1)
df_b.head()
```

```python
df_b.shape
```

```python
df_b.columns
```

missing value를 포함한 col은 전부 drop

```python
df_b = df_b.drop(['PSG_date', 'PSG_ID', 'Final Diagnosis'], axis=1)
df_b.to_csv('../data/survey_psg_data_b.csv', index=False)
```

```python
df_a = df.dropna(axis=1, how='all')
df_a.head()
```

```python
df_a.shape
```

```python
missings = df_a.isnull().sum()
missings_cols = []
for col, value in missings.iteritems():
    if value != 0:
        print(col, "\t", value)
        missings_cols.append(col)
```

```python
df_a = df_a.drop('수면제이름', axis=1)
df_a.head()
```

```python
idxs_list = []
for col in df_a.columns:
    if col == 'ID':
        continue
    na_row = df_a[df_a[col] == 999]
    if len(na_row) != 0:
        idxs_list.append(col)
        print(col, '\t', len(na_row))
```
