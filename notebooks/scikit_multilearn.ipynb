```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../data/preprocessed_data_a.csv')
data.head()
```

```python
from sklearn.model_selection import train_test_split

target = data.iloc[:, 2:10]
features = data.drop(data.iloc[:, 2:10], axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
y_train_converted = y_train.apply(lambda y: y.argmax(), axis=1).values
```

```python
from sklearn.ensemble import RandomForestClassifier
from skmultilearn.ensemble import RakelD

classifier = RakelD(
    base_classifier=RandomForestClassifier(),
    base_classifier_require_dense=[True, True],
    labelset_size=3
)

classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
parameters = {
    'base_classifier_require_dense': [[True, True], [True, False], [False, False], [False, True]],
    'labelset_size': list(range(1, 11))
}
score = 'f1_samples'

clf = GridSearchCV(RakelD(base_classifier=RandomForestClassifier()), parameters, scoring=score)
clf.fit(X_train, y_train)

print(clf.best_params_, clf.best_score_)
```

```python
parameters = {
    'base_classifier_require_dense': [[True, True], [True, False], [False, False], [False, True]],
    'labelset_size': list(range(1, 11)),
    'model_count': list(range(1, 11))
}
score = 'f1_samples'

clf = GridSearchCV(RakelO(base_classifier=RandomForestClassifier()), parameters, scoring=score)
clf.fit(X_train, y_train)

print(clf.best_params_, clf.best_score_)
```

```python
from sklearn.metrics import *

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
from skmultilearn.ensemble import RakelO

classifier = RakelO(
    base_classifier=RandomForestClassifier(),
    base_classifier_require_dense=[True, True],
    labelset_size=y_train.shape[1] // 4,
    model_count=6
)

classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
from skmultilearn.adapt import BRkNNbClassifier
from sklearn.model_selection import GridSearchCV

X = X_train.to_numpy()
y = y_train.to_numpy()

parameters = {'k': range(1,10)}
score = 'f1_macro'

clf = GridSearchCV(BRkNNbClassifier(), parameters, scoring=score)
clf.fit(X, y)

print(clf.best_params_, clf.best_score_)
```

```python
from skmultilearn.adapt import MLTSVM

parameters = {'c_k': [2**i for i in range(-5, 5, 2)]}
score = 'f1_macro'

X_matrix = np.asmatrix(X)
y_matrix = np.asmatrix(y)

clf = GridSearchCV(MLTSVM(), parameters, scoring=score)
clf.fit(X_matrix, y_matrix)

print(clf.best_params_, clf.best_score_)
```
