```python
df = df.drop(['수면제이름', '검사당일\n음주량'], axis=1)
print(df.shape)
df.head()
```

# Data A

```python
df_a = df.dropna()
print(df_a.shape)
df_a.head()
```

```python
cols_list = []
for col in df_a.columns:
    if col == 'ID':
        continue
    na_row = df_a[df_a[col] == 999]
    if len(na_row) != 0:
        cols_list.append(col)
        print(col, '\t', len(na_row))
```

```python
df_a = df_a.drop(['HADS-A', 'HADS-D', 'WHO-QOL'], axis=1)
print(df_a.shape)
df_a.head()
```

```python
cols_list.remove('HADS-A')
cols_list.remove('HADS-D')
cols_list.remove('WHO-QOL')
```

```python
idxs_list = []
for col in cols_list:
    idxs_list += list(df_a.loc[df_a[col] == 999].index)

idxs_list = list(set(idxs_list))
print(idxs_list)
```

```python
df_a = df_a.drop(idxs_list)
print(df_a.shape)
df_a.head()
```

```python
df_a = df_a.drop(['ID', 'PSG_date', 'PSG_ID', 'Final Diagnosis'], axis=1)
print(df_a.shape)
df_a.head()
```

```python
df_a.to_csv('../data/isa_data_a.csv', index=False)
```

```python
missings = df_a.isnull().sum()
for col, value in missings.iteritems():
    if value != 0:
        print(col, "\t", value)
```

```python
for col in df_a.columns:
    na_row = df_a[df_a[col] == 999]
    if len(na_row) != 0:
        print(col, '\t', len(na_row))
```
